```python
import numpy as np
import cv2
import time

cap = cv2.VideoCapture(0)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
fgbg = cv2.createBackgroundSubtractorKNN()
# Инициализация переменных для расчета FPS
frame_count = 0
start_time = time.time()

cntr = 0
while True:
    cntr += 1
    if cntr == 100:
        fgbg = cv2.createBackgroundSubtractorKNN()
        cntr = 0
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1
    # Текущее время
    current_time = time.time()
    # Время, прошедшее с начала замера
    elapsed_time = current_time - start_time
    if elapsed_time > 0:  # Исключаем деление на ноль
        fps = frame_count / elapsed_time
        # Выводим FPS на кадре
        cv2.putText(frame, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 10, 20), 2)

    fgmask = fgbg.apply(frame)
    fgmask = cv2.morphologyEx(fgmask, cv2.MORPH_OPEN, kernel)

    # Вместо создания серой копии используем исходное цветное изображение
    colored_frame = frame.copy()
    
    # Зарисовываем движущиеся объекты красным цветом, оставляя фон цветным
    colored_frame[fgmask == 255] = [0, 0, 255]  # BGR
    
    cv2.imshow('Motion Detection', colored_frame)
    
    if cv2.waitKey(1) & 0xFF == 27:  # Нажмите ESC для выхода
        break

cap.release()
cv2.destroyAllWindows()
```
